```python
import requests
import os

def download_github_folder(api_url, local_dir):
    """
    Recursively downloads files and folders from a GitHub repository using the REST API.
    """
    response = requests.get(api_url)
    
    if response.status_code != 200:
        print(f"Error: Could not fetch content from {api_url}")
        return

    items = response.json()

    # Ensure the local directory exists
    if not os.path.exists(local_dir):
        os.makedirs(local_dir)
        print(f"Created directory: {local_dir}")

    for item in items:
        if item['type'] == 'file':
            # Download individual files
            file_url = item['download_url']
            file_path = os.path.join(local_dir, item['name'])
            
            print(f"Downloading file: {item['name']}...")
            file_data = requests.get(file_url).content
            with open(file_path, 'wb') as f:
                f.write(file_data)
                
        elif item['type'] == 'dir':
            # Create sub-directory and go deeper (Recursion)
            new_api_url = item['url']
            new_local_dir = os.path.join(local_dir, item['name'])
            print(f"Entering folder: {item['name']}...")
            download_github_folder(new_api_url, new_local_dir)

# --- CONFIGURATION ---
# Target: https://github.com/hereandnowai/ml-data-hereandnowai/tree/main/data-for-eda
repo_owner = "hereandnowai"
repo_name = "ml-data-hereandnowai"
folder_path = "data-for-eda" 

# GitHub API URL format: https://api.github.com/repos/{owner}/{repo}/contents/{path}
api_base_url = f"https://api.github.com/repos/{repo_owner}/{repo_name}/contents/{folder_path}"

# Start the download process into a local folder named 'data'
print("Starting bulk download from HERE AND NOW AI GitHub...")
download_github_folder(api_base_url, "data")
print("\nAll files and folders downloaded successfully!")
```
