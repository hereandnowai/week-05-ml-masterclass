## 1. What is Machine Learning?
Machine Learning is the field of study that gives computers the ability to learn without being explicitly programmed. Instead of hard-coding rules, we provide data to an algorithm, which then builds a model to make predictions or decisions.

**Real-world Example:** 
Predicting the price of a house. Instead of writing complex rules for every city and square foot, we show the computer 10,000 past sales, and it learns the relationship between "Size" and "Price" automatically.

---

## 2. AI vs ML vs DL
While often used interchangeably, these represent a nested hierarchy:

*   **Artificial Intelligence (AI):** The broad field of creating systems that can perform tasks requiring human-like intelligence (e.g., Playing Chess, Logic systems).
*   **Machine Learning (ML):** A subset of AI focused on algorithms that improve performance as they are exposed to more data (Learning from experience).
*   **Deep Learning (DL):** A subset of ML inspired by the structure of the human brain (Neural Networks). It excels at multi-layered patterns like Voice Recognition and Image Detection.

**The Analogy:**
*   **AI** is the Brain.
*   **ML** is the process of learning from past experiences.
*   **DL** is deep, specialized learning for complex tasks like recognizing a face in a crowd.

````python
# 4. Simple ML Demo: Linear Regression
# Predicting a simple output 'y' from a single input 'x'
import numpy as np
from sklearn.linear_model import LinearRegression

# Small dataset: Area (x) and Prices (y)
x = np.array([[500], [700], [1000], [1200]]) # Sqft
y = np.array([250000, 350000, 500000, 600000]) # Price in USD

# Train (Fit) the model
model = LinearRegression()
model.fit(x, y)

# Predict for a new Area (900 Sqft)
prediction = model.predict([[900]])
print(f"Predicted Price for 900 Sqft: ${prediction[0]:,.2f}")
````

## 5. Output Interpretation
In the above code, the model learned a **Linear Relationship** ($y = mx + b$). 
1.  **Learned Slope ($m$):** The price increase per square foot.
2.  **Learned Intercept ($b$):** The base price.

Because the model found a straight-line pattern in the history (500, 700, 1000, 1200), it can now accurately estimate the price for any square footage within that range.

## 3. Cosine Similarity in Practice
Let's build a small genre-preference matrix for 4 users.

````python
data = {
    'Action': [5, 4, 1, 1],
    'Sci-Fi': [4, 5, 0, 1],
    'Romance': [1, 0, 5, 4],
    'Comedy': [2, 1, 4, 5]
}
users = ['Zenil', 'Mihir', 'Avishka', 'Aditi']
df_ratings = pd.DataFrame(data, index=users)

similarity = cosine_similarity(df_ratings)
df_sim = pd.DataFrame(similarity, index=users, columns=users)

plt.figure(figsize=(8, 6))
sns.heatmap(df_sim, annot=True, cmap='coolwarm')
plt.title("Recommendation Matrix (User Similarity)")
plt.show()
````

## 6. Types of Machine Learning
There are four primary ways a machine can learn from data:

### Supervised Learning
The model is trained with **Labeled Data**. It knows the input ($x$) and its correct output ($y$).
*   **Example:** House price prediction (Input: House Sqft → Output: Price).

### Unsupervised Learning
The model finds hidden patterns in **Unlabeled Data**. It doesn't know the "correct" results beforehand.
*   **Example:** Customer segmentation (Grouping shoppers by age and income).

### Semi-Supervised Learning
A mix of labeled and unlabeled data. Typically, a small amount of labeled data is combined with a huge amount of raw, unlabeled data.

### Reinforcement Learning
Learning through **Trial and Error**. The model (Agent) performs actions and receives rewards or penalties based on the outcome.
*   **Example:** Self-driving cars OR a Chess-playing AI.

---

## 7. Quick Comparison Table
| Type | Data Status | Common Use-Case |
| :--- | :--- | :--- |
| **Supervised** | Labeled ($X, Y$) | Regression, Classification |
| **Unsupervised** | Unlabeled ($X$ only) | Clustering, Dimensionality Reduction |
| **Semi-Supervised** | Partially Labeled | Large-scale Web Data |
| **Reinforcement** | Trial & Error (Rewards) | Gaming, Robotics |

## 4. Decision Trees & Rule-Based Learning
Decision Trees are one of the most interpretable ML models.

````python
iris = load_iris()
X, y = iris.data, iris.target

dt = DecisionTreeClassifier(max_depth=3, criterion='entropy')
dt.fit(X, y)

plt.figure(figsize=(12, 8))
plot_tree(dt, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
````
